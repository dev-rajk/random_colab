```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

```

```python
# Load your data
df = pd.read_csv('/content/Ghy_avg_temp_51yrs - ghy_date_avg_73-23.csv')



# Remove February 29th from the dataset
df = df[~df['datetime'].str.contains('29-02')]

# Interpolate missing values
df.interpolate(method='linear', inplace=True, axis=0)

# Convert the 'datetime' column to a numerical format for plotting
df['datetime'] = pd.to_datetime(df['datetime'], format='%d-%m').dt.dayofyear

# Prepare the data for Plotly
X = df['datetime'].values
Y = np.array(df.columns[1:], dtype=int)
Z = df.iloc[:, 1:].values
```

```python
# Create a 3D surface plot
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y, colorscale='Viridis')])

# Add title and labels
fig.update_layout(title='3D Plot of Average Temperature Across 51 Years',
                  scene=dict(xaxis_title='Date (Day of Year)',
                             yaxis_title='Year',
                             zaxis_title='Average Temperature (Â°C)'),
                  autosize=True,
                  width=800, height=800)

# Show plot
fig.show()
```
