```python
import imdlib as imd
import pandas as pd
import numpy as np
```

```python
start_yr = 1965
end_yr = 2023
variable = 'tmax' # other options are ('tmin'/ 'tmax'/ 'rain')
data = imd.get_data(variable, start_yr, end_yr, fn_format='yearwise')
```

```python
lat = 26.127023
lon = 91.714114
data.to_csv('test2.csv', lat, lon)

```

```python
df_rain = pd.read_csv("/content/test2_26.13_91.71.csv")
df_rain['DateTime'] = pd.to_datetime(df_rain['DateTime'])
df_rain['day_month'] = df_rain['DateTime'].dt.strftime('%m-%d')
df_rain['year'] = df_rain['DateTime'].dt.year
pivot_df = df_rain.pivot_table(index='day_month', columns='year', values='26.127023 91.714114')
pivot_df.reset_index(inplace=True)
# pivot_df.replace(0, np.nan, inplace=True)
pivot_df.to_csv('tmax.csv', index=False)
```
